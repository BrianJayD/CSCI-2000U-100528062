```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

data = "Source_data_for_CFR_vaccine_map_abridged.csv"
categories = np.genfromtxt(data, dtype=str, delimiter=',', usecols=0, skip_header=1)
latitude = np.genfromtxt(data, dtype=float, delimiter=',', usecols=1, skip_header=1)
longitude = np.genfromtxt(data, dtype=float, delimiter=',', usecols=2, skip_header=1)
cases = np.genfromtxt(data, dtype=int, delimiter=',', usecols=3, skip_header=1)

#Testing to see if inputed into arrays
#print (categories)
#print (latitude)
#print (longitude)
#print (cases)

dataLength = len(categories)

#Measles data
lMeasles = 0
for i in range(dataLength):
    if categories[i] == "Measles":
        if cases[i] > lMeasles:
            lMeasles = cases[i]

map = Basemap(projection='mill', lat_0=0, lon_0=0)
map.drawmapboundary(fill_color='aqua')
map.fillcontinents(color='green',lake_color='aqua')
map.drawcoastlines()

for i in range(dataLength):
    if categories[i] == "Measles":
        x,y = map(longitude[i], latitude[i])
        sizeOfMarker = ((cases[i]/lMeasles)*30)+5
        map.plot(x, y, 'ro', markersize=sizeOfMarker, alpha=0.5)

plt.title("Measels Outbreak 100528062")
plt.show()
```
